```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data=sns.load_dataset("tips")
```

```python
data
```

```python
sns.catplot(x="day",y="total_bill",data=data,kind='swarm')
plt.show()
```

```python
sns.catplot(x="day",y="total_bill",data=data,kind='box')
plt.show()
```

```python
sns.catplot(x="day",y="total_bill",data=data,kind='bar')
plt.show()
```

```python
sns.catplot(x="day",y="total_bill",data=data,kind='violin')
plt.show()
```

```python
sns.catplot(x="day",y="total_bill",data=data,kind='boxen')
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.countplot(data['day'])
```

```python
sns.boxplot(data['total_bill'])
plt.show()
```
